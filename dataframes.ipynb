## Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution
vis_1 = sns.distplot(df["InternetUsers"])
```

```python
#Boxplots
vis_2 = sns.boxplot(data = df, x = "IncomeGroup", y = "BirthRate")
```

```python
#Linear model plot

vis_3 = sns.lmplot(data = df, x= "InternetUsers", y = "BirthRate", fit_reg = False, hue = 'IncomeGroup', size  =10)
```

```python
#To increase the size of the markers, use argument scatter_kws by passsing dictionary values in it
vis_3 = sns.lmplot(data = df, x= "InternetUsers", y = "BirthRate", fit_reg = False, hue = 'IncomeGroup', size  =10, \
                   scatter_kws = {"s": 100})
```
